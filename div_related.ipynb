```python
import json
import operator
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
from nltk import Tree
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
f = open('org.txt',"r")
text = f.read()
org_list = text.split(',')
for i in range(len(org_list)):
    while org_list[i][0] == ' ':
        org_list[i] = org_list[i][1:]
org_list[-1] = org_list[-1][:-1]
for i in range(len(org_list)):
    org_list[i] = org_list[i].lower()
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
stem_list = []
for word in org_list:
    stem = porter_stemmer.stem(word)
    stem_list.append(stem)
```

```python
org_list.extend(stem_list)
```

```python
word_list = []
keywords = open('keywords.txt',"r")
for line in keywords:
    w = line
    if w[-1] == '\n':
        w = w[:-1]
    word_list.append(w)
    stem = porter_stemmer.stem(w)
    word_list.append(stem)
len(word_list)
```

```python
div_list = ['diverse','inclusion','equal','fair','unbias','integrat','open','broaden','differ','involvement','collective','vary','multiple','affirmation','active','constitent','non-discriminating','everyone','free','various']
stem_list = []
for word in div_list:
    stem = porter_stemmer.stem(word)
    stem_list.append(stem)
div_list.extend(stem_list)
```

```python
div_set = set(div_list)
org_set = set(org_list)
word_set = set(word_list)
```
