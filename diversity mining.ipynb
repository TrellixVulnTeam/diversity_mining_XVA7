```python
import json
import operator
import numpy as np
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
f = open("wal.txt","r") 
text = f.read()
print (text)
```

```python
from nltk.tokenize import sent_tokenize
sent_list = sent_tokenize(text)
len(sent_list)
```

```python
for i in range(10):
    print (sent_list[i])
```

```python
sent_tagged = []
for sent in sent_list:
    words = nltk.word_tokenize(sent)
    sent_tagged.append(nltk.pos_tag(words))
    
```

```python
sent_tagged[-10:]
```

```python
#grammar = "NP: {<DT>?<JJ>*<NN>}"
#grammat = "NP: {(<V\w+>|<NN\w?>)+.*<NN\w?>}"
grammar = r"""
  NP: {<DT|JJ|NN.*>+}          # Chunk sequences of DT, JJ, NN
  PP: {<IN><NP>}               # Chunk prepositions followed by NP
  VP: {<VB.*><NP|PP|CLAUSE>+$} # Chunk verbs and their arguments
  CLAUSE: {<NP><VP>}           # Chunk NP, VP
  """
#cp = nltk.RegexpParser(grammar)
chunkParser = nltk.RegexpParser(grammar)
tree = chunkParser.parse(sent_tagged[0])
```

```python
print (tree)
```

```python
phrases = [chunkParser.parse(sublist) for sublist in sent_tagged]

leaves = [[subtree.leaves() for subtree in tree.subtrees(filter = lambda t: t.label == 'NP')] for tree in phrases]
```

```python
continuous_chunk = []
current_chunk = []

for t in phrases:
    for subtree in t:
        if type(subtree) == Tree:
            current_chunk.append(" ".join([token for token, pos in subtree.leaves()]))
        elif current_chunk:
            named_entity = " ".join(current_chunk)
            if named_entity not in continuous_chunk:
                continuous_chunk.append(named_entity)
                current_chunk = []
        else:
            continue
```
